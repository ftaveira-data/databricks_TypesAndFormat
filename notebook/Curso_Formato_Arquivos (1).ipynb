```
# Lendo o arquivo json

df_01 = spark.read.json("/FileStore/tables/tipos_formatos/arquivos_curso/PNSB.json")
display(df_01)
```

```
# renomear as colunas 

df_01 = df_01.withColumnRenamed("D1C", "cod_regiao")\
    .withColumnRenamed("D1N", "regiao")\
        .withColumnRenamed("D2C", "cod_variavel")\
            .withColumnRenamed("D2N", "variavel")\
                .withColumnRenamed("D3C", "cod_ano")\
                    .withColumnRenamed("D3N", "ano")\
                        .withColumnRenamed("D4C", "cod_doenca") \
                            .withColumnRenamed("D4N", "doenca")\
                                .withColumnRenamed("MC", "cod_medida")\
                                    .withColumnRenamed("MN", "medida")\
                                        .withColumnRenamed("NC", "cod_nivel_territorial")\
                                            .withColumnRenamed("NN", "nivel_territorial")\
                                                .withColumnRenamed("V", "valor")

display(df_01)
```

```
# retirar a primeira linha

df_01 = df_01.filter(df_01.valor!="Valor")
display(df_01)
```

```
# visualiazar a tipagem

df_01.printSchema()
```

```
# ajustar os tipos 

from pyspark.sql.functions import col
from pyspark.sql.types import IntegerType

df_01_new = df_01.withColumn ("cod_regiao", col("cod_regiao").cast (IntegerType()))\
    .withColumn("cod_variavel", col("cod_variavel").cast (IntegerType()))\
        .withColumn("cod_ano", col("cod_ano").cast (IntegerType()))\
            .withColumn("ano", col("ano").cast (IntegerType()))\
                .withColumn("cod_doenca", col("cod_doenca").cast (IntegerType()))\
                    .withColumn("cod_medida", col("cod_medida").cast (IntegerType()))\
                        .withColumn("cod_nivel_territorial", col("cod_nivel_territorial").cast (IntegerType()))\
                            .withColumn("valor", col("valor").cast (IntegerType()))
```

```
df_01_new.printSchema()
```

```
# escrevendo um arquivo json e comprimindo o arquivo 

display(dbutils.fs.ls("/FileStore/tables/tipos_formatos/arquivos_curso"))
```

```
# comprimindo o arquivo 

df_01_new.write.option("compression", "gzip").mode("overwrite").format("json").save("/FileStore/tables/tipos_formatos/arquivos_curso/json_gzip/")
```

```
from pyspark.sql import SparkSession
from pyspark.sql.functions import lit

# Iniciar a sessão Spark
spark = SparkSession.builder.appName("fileSizeComparison").getOrCreate()

# Listas dos caminhos
paths = [
    "/FileStore/tables/tipos_formatos/arquivos_curso/json_gzip",
    "/FileStore/tables/tipos_formatos/arquivos_curso/pnsb_avro_snappy",
    "/FileStore/tables/tipos_formatos/arquivos_curso/pnsb_csv",
    "/FileStore/tables/tipos_formatos/arquivos_curso/pnsb_orc_zlib",
    "/FileStore/tables/tipos_formatos/arquivos_curso/pnsb_parquet_gzip"
]

# Lista para armazenar os resultados
results = []

# Iterar sobre cada caminho e calcular o tamanho total
for path in paths:
    size = sum([item.size for item in dbutils.fs.ls(path)])
    results.append((path, size))

# Criar DataFrame com os resultados
df_sizes = spark.createDataFrame(results, ["Path", "Size"])

# Mostrar o DataFrame
display(df_sizes)
```
